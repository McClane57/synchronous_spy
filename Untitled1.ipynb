```
import numpy as np
import time
from random import randint, sample
from sklearn.datasets import load_svmlight_file
import numpy.linalg as lg
import datetime
from scipy.optimize import minimize

from projections_l1 import *
from functions import *
from sklearn.datasets import load_svmlight_file
```

Initialization of data


```
dataset_url = "./full_rcv1_train.dat"

data = load_svmlight_file(dataset_url)

A = csr_matrix(data[0])
b = csr_matrix(data[1]).T
m, n = A.shape
```

Data partitioning

```
M = 20 #amount of workers

m_w = int(np.floor(m/M))

A_w = []
b_w = []

for i in range(19):
  index_start = m_w*i
  index_end = m_w*(i+1)
  A_w.append(A[index_start:index_end,:])
  b_w.append(b[index_start:index_end])

A_w.append(A[19*m_w:,:])
b_w.append(b[19*m_w:])

weight = []
for i in range(20):
  weight.append(b_w[i].shape[0]/(m+0.0))
```

Constants initialization

```
Lambda2 = 0.0
Lambda1 = 0.001

L = 0.25 + Lambda2
mu = Lambda2

gam = 2./(mu+L)

COMM_MAX = 3000*n
ITE_MAX = 1000

def function(x):
  return F_l1(csr_matrix(x).T, A, b, Lambda2, Lambda1)
```

Runs initialization

```
amount_of_coordinates = [10,100,1000,5000, 'full']
starting_point = np.random.randn(n)
```

Run


```

for run in range(len(amount_of_coordinates)):
  barx = np.copy(starting_point)
  sp_tab = []
  f_tab = []
  ite_tab = []
  comm_tab = []

  x_w = []
  for i in range(20):
    x_w.append(np.copy(barx))
  
  ite = 1
  comm = 0

  while (ite < ITE_MAX):
  #while (comm < COMM_MAX):
    S_k = []
    if ite%10 == 1:
      print(ite)
      x_k = prox_l1(csr_matrix(barx).T, gam, Lambda1).T.toarray()[0]
      sp_tab.append(len(np.flatnonzero(x_k)) * 100.0 / float(n))
      f_tab.append(function(x_k))
      print(f_tab[-1])
      ite_tab.append(ite)
      comm_tab.append(comm)

    comm += 20*len(list(np.flatnonzero(x_k)))
    if amount_of_coordinates[run] != 'full':
      S_k = list(np.flatnonzero(x_k))

      S_bar = AdjointSetL1(n, S_k)
      n_bar = min(len(S_bar), amount_of_coordinates[run])
      S_k = S_k+list(np.random.choice(S_bar, n_bar, replace = False))

    for i in range(20):
      if amount_of_coordinates[run] != 'full':
        x_w[i][S_k] = x_k[S_k] - gam * (f_grad(csr_matrix(x_k).T, A_w[i], b_w[i], Lambda2).T.toarray()[0][S_k])
        comm += len(list(np.flatnonzero(x_w[i][S_k])))
      else:
        x_w[i] = x_k - gam * (f_grad(csr_matrix(x_k).T, A_w[i], b_w[i], Lambda2).T.toarray()[0])
        comm += len(list(np.flatnonzero(x_w[i])))                       
    barx = np.zeros(n)
    for i in range(20):
      barx += weight[i]*x_w[i]
    ite += 1

  out_file_url = "rcv1_train_20w_nonstrongly"+str(amount_of_coordinates[run]) + ".dat"
  out_file = open(out_file_url, 'w')

  line = "Ite Exchanges Sparsity Functional_value\n"
  print(line)
  out_file.write(line)

  for i in range(len(sp_tab)):
      line = "{} {} {} {}\n".format(ite_tab[i], comm_tab[i], sp_tab[i], f_tab[i])
      out_file.write(line)
      print(line)
  out_file.close()



```

Figures


```
import matplotlib.pyplot as plt
import pandas as pd

runs = []
legend = []
markers = []
lines = []
colors = ['y', 'g', 'b', 'k', 'm']
for run in range(len(amount_of_coordinates)):
  runs.append(pd.read_csv("rcv1_train_20w_nonstrongly"+str(amount_of_coordinates[run]) + ".dat", delim_whitespace=True))
  if amount_of_coordinates[run] != 'full':
    legend.append("SPY_"+str(amount_of_coordinates[run]))
    markers.append("$"+str(amount_of_coordinates[run])+"$")
    lines.append(':')
  else:
    legend.append("PGD")
    markers.append("")
    lines.append('-')
```

```
Fmin = 10.
for i in range(len(runs)):
  Fmin = min(Fmin, min(runs[i]['Functional_value'] - 1e-15))
```

Sparsity

```
for i in range(len(runs)):
    plt.plot(runs[i]['Ite'],runs[i]['Sparsity'], lines[i],marker = markers[i], markevery = (10 + 20*i,10)	, color = colors[i%5], label = legend[i], markersize = 10 )
plt.xlabel("Iteration",fontsize = 'x-large')
plt.ylabel("Support (%)",fontsize = 'x-large')
plt.legend()
plt.yscale('log')
plt.savefig("rcv1_train_20w_0001_00001_density.pdf")
plt.show()
```

Functional suboptimality versus iteration

```
plt.figure()
for i in range(len(runs)):
    plt.semilogy(runs[i]['Ite'],runs[i]['Functional_value'] - Fmin, lines[i], marker = markers[i], markevery = (10 + 20*i,10), color = colors[i%5], label =  legend[i], markersize = 10)
plt.xlabel("Iteration",fontsize = 'x-large')
plt.ylabel("Suboptimality",fontsize = 'x-large')
plt.legend()
plt.savefig("rcv1_train_20w_0001_00001_fun_vs_ite_log_coordinate.pdf")
plt.show()
```

```
plt.figure()
for i in range(len(runs)):
    plt.semilogy(runs[i]['Exchanges'],runs[i]['Functional_value'] - Fmin, lines[i], marker = markers[i], markevery = (10 + 20*i,10), color = colors[i%5], label =  legend[i], markersize = 10)
plt.xlabel("Exchanges",fontsize = 'x-large')
plt.ylabel("Suboptimality",fontsize = 'x-large')
plt.legend()
plt.savefig("rcv1_train_20w_0001_00001_fun_vs_ex_log_coordinate.pdf")
plt.show()
```
